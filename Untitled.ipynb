```python
import sys
sys.path.append("c:/users/david/desktop/sandbox")
import pandas as pd
from pyrecs.train import lightfm_wrapper

%load_ext autoreload
%autoreload 2
```

```python
df = pd.DataFrame({'users':['user1','user1','user1','user1','user2'], 
                   'items':['item1','item1','item3','item1','item2']})
df
```

```python
lfm = lightfm_wrapper.LightFM(rows_col='users', columns_col='items', interactions_type='ones')
lfm.preprocess(df)
```

```python
model_kwargs = {
    'no_components':10,
    'learning_rate':0.05,
    'loss':'warp',
    'random_state':42
}
train_kwargs = {
  'num_epochs':10,
  'user_features':None,
  'item_features':None,
  'num_threads':1
}
truth = {
    'user1':['item1', 'item2', 'item3'],
    'user2':['item1', 'item2', 'item3']
}
# TODO: Need to update lookup for if truth has items that are not in train and are therefore not in the lookups
lfm.train(model_kwargs=model_kwargs,
          train_kwargs=train_kwargs,
          test_dict=truth,
          n_recs=3)
```
