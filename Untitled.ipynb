```python
import sys
sys.path.append("c:/users/david/desktop/sandbox")
import pandas as pd
from pyrecs.train import lightfm_wrapper

%load_ext autoreload
%autoreload 2
```

```python
df = pd.DataFrame({'users':['user1','user1','user1','user1','user2'], 
                   'items':['item1','item1','item3','item1','item2']})
df
```

```python
lfm = lightfm_wrapper.LightFM(rows_col='users', columns_col='items', interactions_type='ones')
lfm.preprocess(df)
```

```python
model_kwargs = {
    'no_components':10,
    'learning_rate':0.05,
    'loss':'warp',
    'random_state':42
}
train_kwargs = {
  'num_epochs':10,
  'user_features':None,
  'item_features':None,
  'num_threads':1
}
truth = pd.DataFrame({'users':['user1','user1','user1','user2', 'user2', 'user2','user2','new_user3'], 
                      'items':['item1','item2','item3','item1', 'item2', 'item3','nnnn','item1']})

lfm.train(model_kwargs=model_kwargs,
          train_kwargs=train_kwargs,
          test_df=truth,
          n_recs=3)
```
